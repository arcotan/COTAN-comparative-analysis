```python
import pandas as pd
```

```python
PVAL_THRESHOLD = 0.001
ABSLFC_THRESHOLD = 2 # TODO: to set
DIR = './data/'
DATASET_NAMES = ['PBMC1', 'PBMC2', 'PBMC3', 'PBMC4']
TOOLS = ['COTAN', 'seurat', 'monocle', 'scanpy', 'scvi-tools']
```

```python
for dataset_name in DATASET_NAMES:
    print('Dataset:', dataset_name)
    markers_size_df = pd.DataFrame()
    for tool in TOOLS:
        if tool == 'scvi-tools':
            file = DIR + dataset_name + '/' + tool + '/antibody/markers_abslfc.csv'
        else:
            file = DIR + dataset_name + '/' + tool + '/antibody/markers_pval.csv'
        markers_df = pd.read_csv(file)
        counts = markers_df.groupby('cluster').size()
        if tool == 'scvi-tools':
            sign_markers_df = markers_df[markers_df['abs_lfc_mean'] > ABSLFC_THRESHOLD]
        else:
            sign_markers_df = markers_df[markers_df['p_val'] < PVAL_THRESHOLD]
        sign_markers_df = sign_markers_df.groupby('cluster').size()
        markers_size_df[tool+' significant markers'] = sign_markers_df
        markers_size_df[tool+' total markers'] = counts
    display(markers_size_df)
```
